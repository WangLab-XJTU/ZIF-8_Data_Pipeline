```python
import pandas as pd 

from toolkit import ReaderFulltext
from toolkit import Extractor
from toolkit import FilePath
from toolkit import Logger
```

```python
filepath = FilePath()
extractor = Extractor()
logger = Logger()
```

```python
#filepath.make_matedata()
count = 0
while True:
    # if count >= 100:break
    count += 1

    doi,journal,file =  filepath.get_next_task()
    if doi is None and journal is None and file is None:
        print('Tasks Completed!')
        break

    doc = ReaderFulltext(file,journal)
    if doc.elements is None:
        filepath._updata_matedata(doi)
        logger.info(f'Fulltext is None, {doi} {journal}')
        continue

    elif doc.paragraphs is None:
        filepath._updata_matedata(doi)
        logger.info(f'Not Found paragraphs from {doi} {journal}')
        continue
    elif doc.targeted_para is None:
        filepath._updata_matedata(doi)
        logger.info(f'Not match target paragraphs from {doi} {journal}')
        continue
    filepath.verify_file_writer(doi, '\n'.join(doc.elements),f'Fulltext_clean.text')
    filepath.verify_file_writer(doi, '\n'.join(doc.paragraphs),f'paragraphs_clean.text')
    filepath.verify_file_writer(doi, '\n'.join(doc.targeted_para),f'target_paragraphes.text')
    
    response = extractor.extract_data('\n'.join(doc.targeted_para))
    filepath.save_extract_data(doi,response)
    print(f'{count}\t{doi} {journal}\tDone')
    
    
```
